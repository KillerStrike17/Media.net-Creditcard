```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
from torch.utils.data import Dataset, DataLoader
from CreditCard.utils import Encoder
```

```python
df = pd.read_csv("credit_customers.csv",encoding = "utf-8")
df.head()
```

## Dataset creation

```python
# label = le.fit_transform(df['checking_status'])
columns = ['checking_status','credit_history','purpose','savings_status','employment','personal_status',
           'other_parties','property_magnitude','other_payment_plans','housing','job','own_telephone',
           'foreign_worker','class']
```

```python
encoded_df = Encoder(df, columns)
```

```python
encoded_df
```

```python
from collections import Counter
Counter(encoded_df['class'])
```

```python
X = encoded_df.iloc[:,:-1]
y = encoded_df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

```python
X_train.head()
```

### Training

```python
from torch.utils.tensorboard import SummaryWriter
%reload_ext tensorboard
```

```python
from CreditCard.model import BinModel
from imblearn.under_sampling import RandomUnderSampler
import numpy as np
from sklearn.model_selection import StratifiedKFold
writer = SummaryWriter()
```

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True)
cv_scores = []
fold = 1
for train, test in kfold.split(X_train, y_train):
    # create model, train, and get accuracy
    print("Fold:",fold)
    model = BinModel()
    under = RandomUnderSampler()
    X_training, y_training = under.fit_resample( X.iloc[train], y.iloc[train])
    X_testing, y_testing = under.fit_resample( X.iloc[test], y.iloc[test])
    # print(y_training)
    acc = model.model_train(model,
                      torch.tensor(X_training.values).to(torch.float32),
                      torch.tensor(y_training).view(-1, 1).to(torch.float32),
                      torch.tensor(X_testing.values).to(torch.float32),
                      torch.tensor(y_testing).view(-1, 1).to(torch.float32),
                      30,writer,fold
                      )
    print("Accuracy: %.2f" % acc)
    # print(Counter(y.iloc[test].values))
    print(Counter(y_testing))
    cv_scores.append(acc)
    fold+=1

# evaluate the model
acc = np.mean(cv_scores)
std = np.std(cv_scores)
print("Model accuracy: %.2f%% (+/- %.2f%%)" % (acc*100, std*100))
```

```python
y_pred = model(torch.tensor(X_test.values).to(torch.float32))
acc = (y_pred.round().detach().numpy().flatten() == y_test).mean()
acc = float(acc)
```
